# Clustering

```python
from sklearn.cluster import KMeans
```

```python
labels=kmeans.labels_
```

```python
labels
```

```python
iris_df['clusters2']=labels
```

```python
plt.figure(figsize=(10,8))
plt.scatter(iris_df['petal length (cm)'],iris_df['petal width (cm)'],c=iris_df['clusters2'])
plt.xlabel("petal length (cm)")
plt.ylabel("petal width (cm)")
plt.title("petal length vs petal width with clusters")
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Create a range of k values to test
k_values = range(1, 11)

# Initialize a list to store the distortion scores
distortion_scores = []

# Loop through each k value and calculate the distortion score
for k in k_values:
    kmeans = KMeans(n_clusters=k,n_init='auto')
    kmeans.fit(iris_df[['petal length (cm)', 'petal width (cm)']])
    distortion_scores.append(kmeans.inertia_)

# Create the elbow plot
plt.figure(figsize=(8, 6))
plt.plot(k_values, distortion_scores, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Distortion Score (Sum of Squared Errors)')
plt.title('Elbow Plot for Optimal k Value')
plt.show()
```

Based on the elbow plot, the optimal value of k is 3. This means that we can use k=3 as the number of clusters for the KMeans algorithm.
